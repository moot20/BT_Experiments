# Energy Consumption Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Load the CSV file
file_path = './energy_consumption_log.csv'
data = pd.read_csv(file_path)

# Convert Time_elapsed_seconds to seconds
def convert_time_to_seconds(time_str):
    parts = time_str.split(':')
    if len(parts) == 3:
        hours, minutes, seconds = parts
    else:
        hours = 0
        minutes, seconds = parts
    total_seconds = int(hours) * 3600 + int(minutes) * 60 + float(seconds)
    return total_seconds

data['Time_elapsed_seconds'] = data['Time_elapsed_seconds'].apply(convert_time_to_seconds)

# Filter the data for the specified files
svm_py_data = data[data['File'] == 'SVM.py']
svm_r_data = data[data['File'] == 'SVM.R']
rf_py_data = data[data['File'] == 'RF.py']
rf_r_data = data[data['File'] == 'RF.R']
```

```python
# Calculate summary statistics for each group
svm_py_summary = svm_py_data.describe()
svm_r_summary = svm_r_data.describe()
rf_py_summary = rf_py_data.describe()
rf_r_summary = rf_r_data.describe()

# Print summary statistics
print("SVM.py Summary Statistics")
print(svm_py_summary)
print("\nSVM.R Summary Statistics")
print(svm_r_summary)
print("\nRF.py Summary Statistics")
print(rf_py_summary)
print("\nRF.R Summary Statistics")
print(rf_r_summary)
```

```python
# Plot comparison of Energy_pkg_Joules
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.boxplot([svm_py_data['Energy_pkg_Joules'], svm_r_data['Energy_pkg_Joules']],
            tick_labels=['SVM.py', 'SVM.R'])
plt.title('Energieverbrauch in J des CPU-Pakets - Vergleich für SVM')

plt.subplot(1, 2, 2)
plt.boxplot([rf_py_data['Energy_pkg_Joules'], rf_r_data['Energy_pkg_Joules']],
            tick_labels=['RF.py', 'RF.R'])
plt.title('Energieverbrauch in J des CPU-Pakets - Vergleich für Random Forest')

plt.tight_layout()
plt.show()
```

```python
# Plot comparison of Energy_psys_Joules
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.boxplot([svm_py_data['Energy_psys_Joules'], svm_r_data['Energy_psys_Joules']],
            tick_labels=['SVM.py', 'SVM.R'])
plt.title('Energieverbrauch in J der Gesamtplatform - Vergleich für SVM')

plt.subplot(1, 2, 2)
plt.boxplot([rf_py_data['Energy_psys_Joules'], rf_r_data['Energy_psys_Joules']],
            tick_labels=['RF.py', 'RF.R'])
plt.title('Energieverbrauch in J der Gesamtplatform - Vergleich für Random Forest')

plt.tight_layout()
plt.show()
```

```python
# Plot comparison of Time_elapsed_seconds
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.boxplot([svm_py_data['Time_elapsed_seconds'], svm_r_data['Time_elapsed_seconds']],
            tick_labels=['SVM.py', 'SVM.R'])
plt.title('Ausführungszeit in Sekunden - Vergleich für SVM')

plt.subplot(1, 2, 2)
plt.boxplot([rf_py_data['Time_elapsed_seconds'], rf_r_data['Time_elapsed_seconds']],
            tick_labels=['RF.py', 'RF.R'])
plt.title('Ausführungszeit in Sekunden - Vergleich für Random Forest')

plt.tight_layout()
plt.show()
```

```python
# Plot comparison of Memory_used_kb
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.boxplot([svm_py_data['Memory_used_kb'], svm_r_data['Memory_used_kb']],
            tick_labels=['SVM.py', 'SVM.R'])
plt.title('Maximaler Speicherbedarf in KB - Vergleich für SVM')

plt.subplot(1, 2, 2)
plt.boxplot([rf_py_data['Memory_used_kb'], rf_r_data['Memory_used_kb']],
            tick_labels=['RF.py', 'RF.R'])
plt.title('Maximaler Speicherbedarf in KB - Vergleich für Random Forest')

plt.tight_layout()
plt.show()
```
